## Funciones de agrupaciÃ³n

```python
medallistaAnio = deportistaDF \
  .join(resultadoDF, deportistaDF.deportista_id == resultadoDF.deportista_id, "left") \
  .join(juegoDF, juegoDF.juego_id == resultadoDF.juego_id, "left") \
  .join(paisDF, deportistaDF.equipo_id == paisDF.id, "left") \
  .join(eventoDF, eventoDF.evento_id == resultadoDF.evento_id, "left") \
  .join(deporteDF, eventoDF.deporte_id == deporteDF.deporte_id, "left") \
  .select("sigla",
          "anio",
          "medalla",
          eventoDF.nombre.alias("Nombre subdisciplina"),
          deporteDF.deporte.alias("Nombre disciplina"),
          deportistaDF.nombre,
          )
```

```python
medallistaAnio.show()
```

```python
#Descartar deportistas sin medalla
medallistaAnio2 = medallistaAnio.filter(medallistaAnio.medalla != "NA") \
  .sort("anio") \
  .groupBy("sigla","anio","Nombre subdisciplina") \
  .count()
```

```python
medallistaAnio2.printSchema()
```

```python
#Metodo recomendado para hacer agrupaciones
medallistaAnio2.groupBy("sigla", "anio") \
  .agg(_sum("count").alias("Total de medallas"), \
       _avg("count").alias("Medallas promedio")).show()
```

## Â¿QUE ES UN UDF?
Las funciones definidas por el usuario o UDF.
Son la funcionalidad agregada en Spark para definir funciones basadas en columnas las cuales permiten entender las capacidades de Spark al momento de transformar el set de datos.

Este tipo de implementaciones son convenientes cuando tenemos un desarrollo extenso donde hemos identificado la periodicidad de tareas repetitivas como suele ser en pasos de limpieza de datos, transformaciÃ³n o renombrado dinÃ¡mico de columnas.

Por lo anterior es comÃºn encontrar en un proyecto de Spark una librerÃ­a independiente donde existen todas estas funciones agregadas para que los desarrolladores involucrados en el proyecto puedan usarlas a conveniencia.

El uso de UDF no implica que las funciones que podemos crear nativamente con Python, Scala, R o Java no sean Ãºtiles. Una UDF tiene el objetivo de ofrecer un estÃ¡ndar interno en el proyecto que nos encontremos realizando. AdemÃ¡s, en caso de ser necesario, una UDF puede ser modificada con ayuda de decoradores para que sea mÃ¡s extensible en diversos escenarios a los cuales nos podemos enfrentar.

Otro motivo para usar UDF es que en el mÃ³dulo de Spark MLlib, la librerÃ­a nativa de Spark para operaciones de Machine Learning, las UDF juegan un papel vital al momento de hacer transformaciones. Por lo cual tener un uso familiar de estas ampliarÃ¡ considerablemente la curva de aprendizaje de Spark MLlib.

### Registro de funciones para que trabajen de forma nativa

```python
#CREAR UDF PARA QUITAR ESPACIOS Y AGREGAR NULL
from pyspark.sql.functions import udf

def conversionEnteros(valor):
  return int(valor) if len(valor) > 0 else None

#Z es comun al registrar una udf
conversionEnteros_udf = udf(lambda z: conversionEnteros(z), IntegerType())
#Dar de alta UDF

spark.udf.register("conversionEnteros_udf", conversionEnteros_udf)
```

```python
#USAR UDF para la transformaciÃ³n de valores
deportistaErrorDF.select(conversionEnteros_udf("altura") \
.alias("alturaUDF")).show()
```

# Comprendiendo la persistencia y particionado

**Problemas al usar un RDD o DF varias veces:**


*   Spark recomputa el componente y sus dependencias cada vez que se ejecuta una acciÃ³n
*   Es costoso(especialmente en problemas iterativos)





**SoluciÃ³n**
*   Conservar el componente en memoria y/o disco.
*   Metodo cache() o persist() ayudan.
*   En PySpark los datos son almacenados de forma serializada.
